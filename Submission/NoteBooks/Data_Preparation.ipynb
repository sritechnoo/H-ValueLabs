# Imports


```python
import os
import re

import pandas as pd
import numpy as np
```

```python
from keras.preprocessing.text import Tokenizer

from keras.preprocessing.sequence import pad_sequences
```

# Read File

```python
trian_path=os.path.join(os.path.pardir, "DataSet", "train.csv")
trian_path
```

```python
train_df=pd.read_csv(trian_path, index_col = False)
train_df = train_df.apply(lambda x: x.astype(str).str.lower())
train_df
```

```python
train_df[['distractor_1','distractor_2','distractor_3']] = train_df.distractor.str.split(",",n = 2,expand=True) 
train_df
```

```python
train_df.drop(columns=['distractor'], inplace=True)
train_df
```

```python
train_df.info()
```

```python
train_df.dropna(inplace=True)
```

```python
train_df.info()
```

```python
train_df
```

```python
distractor_start = 'ssss '
distractor_end = ' eeee'
```

```python
def ReplaceStartAndEnd(s, START_WORD=distractor_start, END_WORD=distractor_end):
    s=s.strip()
    if (s[0] == s[-1]) and s.startswith(("'", '"')):
        return START_WORD + s[1:-1] + END_WORD
    else:
        return s
```

```python
train_df["distractor_1_data"]=train_df.distractor_1.apply(ReplaceStartAndEnd)
train_df["distractor_2_data"]=train_df.distractor_2.apply(ReplaceStartAndEnd)
train_df["distractor_3_data"]=train_df.distractor_3.apply(ReplaceStartAndEnd)
train_df
```

# Tokenizer

## Input

```python
help(Tokenizer)
```

```python
tokenizer_input = Tokenizer(oov_token='ooov')
sentences_input = pd.concat((train_df["question"],train_df["answer_text"]), ignore_index=True).tolist()
tokenizer_input.fit_on_texts(sentences_input)

voc_size_input =  len(tokenizer_input.word_index) + 1
idx_input = tokenizer_input.word_index
inverse_map_input = dict(zip(idx_input.values(), idx_input.keys()))
```

```python
question_tokens = tokenizer_input.texts_to_sequences(train_df["question"])
answer_text_tokens = tokenizer_input.texts_to_sequences(train_df["answer_text"])
```

```python
train_df["question_tokens"]=question_tokens
train_df["answer_text_tokens"]=answer_text_tokens
```

```python
num_tokens_input = [len(tokens) for tokens in question_tokens + answer_text_tokens]
num_tokens_input = np.array(num_tokens_input)

max_num_tokens_input=int(np.mean(num_tokens_input) + 2 * np.std(num_tokens_input))+1
per_covered_tokens_input = np.sum(num_tokens_input < max_num_tokens_input) / len(num_tokens_input)

len(num_tokens_input), np.min(num_tokens_input), np.max(num_tokens_input), np.mean(num_tokens_input), max_num_tokens_input , per_covered_tokens_input
```

# Output

```python
tokenizer_output = Tokenizer(oov_token='ooov')
sentences_output = pd.concat((train_df["distractor_1_data"], train_df["distractor_2_data"],train_df["distractor_3_data"],), ignore_index=True).tolist()
tokenizer_output.fit_on_texts(sentences_output)


voc_size_output =  len(tokenizer_output.word_index)  + 1
idx_output = tokenizer_output.word_index
inverse_map_output= dict(zip(idx_output.values(), idx_output.keys()))
```

```python
distractor_1_tokens = tokenizer_output.texts_to_sequences(train_df["distractor_1_data"])
distractor_2_tokens = tokenizer_output.texts_to_sequences(train_df["distractor_2_data"])
distractor_3_tokens = tokenizer_output.texts_to_sequences(train_df["distractor_3_data"])
```

```python
train_df["distractor_1_tokens"]=distractor_1_tokens
train_df["distractor_2_tokens"]=distractor_2_tokens
train_df["distractor_3_tokens"]=distractor_3_tokens
```

```python
num_tokens_output = [len(tokens) for tokens in distractor_1_tokens + distractor_2_tokens + distractor_3_tokens]
num_tokens_output = np.array(num_tokens_output)

max_num_tokens_output=int(np.mean(num_tokens_output) + 2 * np.std(num_tokens_output))+1
per_covered_tokens_output = np.sum(num_tokens_output < max_num_tokens_output) / len(num_tokens_output)

len(num_tokens_output), np.min(num_tokens_output), np.max(num_tokens_output), np.mean(num_tokens_output), max_num_tokens_output , per_covered_tokens_output
```

```python
train_df
```

# Padding

## Input

```python
pad_input='pre'

question_tokens_pad = pad_sequences(question_tokens, maxlen=max_num_tokens_input, padding=pad_input, truncating=pad_input)
answer_text_tokens_pad = pad_sequences(answer_text_tokens, maxlen=max_num_tokens_input, padding=pad_input, truncating=pad_input)
```

```python
train_df["question_tokens_pad"]=list(question_tokens_pad)
train_df["answer_text_tokens_pad"]=list(answer_text_tokens_pad)
```

## Output

```python
pad_output='post'

distractor_1_tokens_pad = pad_sequences(distractor_1_tokens, maxlen=max_num_tokens_output, padding=pad_output, truncating=pad_output)
distractor_2_tokens_pad = pad_sequences(distractor_2_tokens, maxlen=max_num_tokens_output, padding=pad_output, truncating=pad_output)
distractor_3_tokens_pad = pad_sequences(distractor_3_tokens, maxlen=max_num_tokens_output, padding=pad_output, truncating=pad_output)
```

```python
train_df["distractor_1_tokens_pad"]=list(distractor_1_tokens_pad)
train_df["distractor_2_tokens_pad"]=list(distractor_2_tokens_pad)
train_df["distractor_3_tokens_pad"]=list(distractor_3_tokens_pad)
```

```python
def tokens_to_string(tokens,inverse_map):
    # Map from tokens back to words.
    words = [inverse_map[token] for token in tokens if token != 0]
    
    # Concatenate all words.
    text = " ".join(words)

    return text
```
